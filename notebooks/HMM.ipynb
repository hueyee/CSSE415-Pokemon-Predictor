```python
import numpy as np
from hmmlearn import hmm
import pandas as pd
```

```python
df = pd.read_parquet("parsed_showdown_replays.parquet")
df.head()
```

```python
df = df.dropna(axis=1)
df = df.drop(['game_id'], axis=1)
```

```python
X = df.drop('p2_revealed_pokemon', axis=1)
y = df['p2_revealed_pokemon']
```

```python
X_number = X.select_dtypes(include='number')
X_categorical = X.select_dtypes(exclude='number')
X_categorical = pd.get_dummies(X_categorical, dtype='int')

X = pd.concat([X_number, X_categorical], axis=1)
```

```python
X.head()
```

```python
print("Unique Pokemon:",len(df['p2_revealed_pokemon'].value_counts()))
```

```python
model = hmm.GaussianHMM(n_components=10, covariance_type="full", n_iter=10, tol=1e-4, random_state=0)
```

```python
model.fit(X)
```

```python
yh = model.predict(X)
```
